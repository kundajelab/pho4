```
def readChromSizes(chrom_sizes_file):
        chrom_size_list = []
        for line in open(chrom_sizes_file):
            (chrom, size) =\
              line.rstrip().split("\t")[0:2]
            chrom_size_list.append((chrom,int(size)))
        return chrom_size_list
```

```
chrms = ["chrI","chrII","chrIII","chrIV","chrV","chrVI","chrVII","chrVIII",
               "chrIX","chrX","chrXI","chrXII","chrXIII","chrXIV","chrXV","chrXVI","chrM"]
```

```
def customChromSizeSort(c):
  return chrms.index(c[0])
```

```
from pyfaidx import Fasta
genome_object = Fasta("sacCer3.genome.fa")

chrom_sizes = readChromSizes("sacCer3.chrom.sizes")
chrom_sizes.sort(key=customChromSizeSort)
```

```
num_chroms = len(chrom_sizes)

fasta_sequences = []
for chrom in chrom_sizes:
  chrom_num = chrom[0]
  chrom_size = chrom[1]
  fasta_sequences.append(genome_object[chrom_num][0:chrom_size].seq)
```

```
import re
p = re.compile("CACGTG")

pho4_motif_list = []
for idx in range(len(fasta_sequences)):
  for m in p.finditer(fasta_sequences[idx].upper()):
    pho4_motif_list.append((chrms[idx],int(m.start())))
```

```
def getCondition(idx):
  condition = ""
  if (idx==0 or idx==1):
    condition = "chip high"
  if (idx==2 or idx==3):
    condition = "chip low"
  if (idx==4 or idx==5):
    condition = "control high"
  if (idx==6 or idx==7):
    condition = "control low"
  return condition
```

```
# do the ranges
region_size = 200
motif_length = 6

range_dict = {}
ranges = []
for idx in range(len(pho4_motif_list)):
  chrom_number = pho4_motif_list[idx][0]
  start = pho4_motif_list[idx][1]-int((region_size/2 - motif_length/2))
  end = pho4_motif_list[idx][1]+int((region_size/2 + motif_length/2-1))
  ranges.append((chrom_number,start,end))
  for num in range(start,end):
    range_dict[chrom_number+str(num)]=0
```

```
# dict with regions to store the counts
range_counts_dict = {}
for idx in range(len(pho4_motif_list)):
  chrom_number = pho4_motif_list[idx][0]
  start = pho4_motif_list[idx][1]-int((region_size/2 - motif_length/2))
  end = pho4_motif_list[idx][1]+int((region_size/2 + motif_length/2-1))
```

```
import pyBigWig

files_list = ["chip_high_pos.bw","chip_high_neg.bw","chip_low_pos.bw","chip_low_neg.bw","control_high_pos.bw","control_high_neg.bw",
               "control_low_pos.bw","control_low_neg.bw"]
  
bigWigs = []

for idx in range(len(files_list)):
  bigWigs.append(pyBigWig.open(files_list[idx]))
        
chrms = ["chrI","chrII","chrIII","chrIV","chrV","chrVI","chrVII","chrVIII",
               "chrIX","chrX","chrXI","chrXII","chrXIII","chrXIV","chrXV","chrXVI"]
        
# retrieve values. 
# counts is a list of entries. entry format is tuple (region,condition,count)
# region is also a tuple. (chrom num, start, end)

#counts_dict = {}
counts = []
for c in chrms:
  for j in range(len(bigWigs)):
    condition = getCondition(j)
    intervals = bigWigs[j].intervals(c)
    for i in intervals:
      start = i[0]
      end = i[1]
      count = i[2]
      start_string = c+str(start)
      end_string = c+str(end)
      if (start_string in range_dict or end_string in range_dict):
        counts.append(((c,start,end),condition,count))
        """
        # DO SOMETHING where you add to dict but you also want to make sure you're distinguishing btwn the different conditions.
        if (start_string in range_dict):
          range_dict[start_string]+=count
        else:
          range_dict[end_string]+=count
          """
      #counts_dict[c+str(start)] = (end,condition,count)
```

```
def getRow(condition):
  row = 0
  if (condition=="chip high"):
    row = 0
  if (condition=="chip low"):
    row = 1
  if (condition=="control high"):
    row = 2
  if (condition=="control low"):
    row = 3
  return row
```

```
import math
index = 0

num_conditions = 4
w, h = len(ranges), num_conditions;
table = [[0 for x in range(w)] for y in range(h)] 

for entry in counts:
  chr_num = entry[0][0]
  start = entry[0][1]
  end = entry[0][2]
  condition = entry[1]
  value = entry[2]
  row = getRow(condition)
  
  if ranges[index][1] <= start <= ranges[index][2] or ranges[index][1] <= end <= ranges[index][2]:
    table[row][index] += abs(value)
  else:
    index=index+1
    index=index%len(ranges)
    table[row][index] += abs(value)
```

```
import numpy as np

# Generate some test data
data = np.array(table)

# Write the array to disk
with open('test.txt', 'w') as outfile:
    # I'm writing a header here just for the sake of readability
    # Any line starting with "#" will be ignored by numpy.loadtxt
    outfile.write('# Array shape: {0}\n'.format(data.shape))

    # Iterating through a ndimensional array produces slices along
    # the last axis. This is equivalent to data[i,:,:] in this case
    for data_slice in data:

        # The formatting string indicates that I'm writing out
        # the values in left-justified columns 7 characters in width
        # with 2 decimal places.  
        np.savetxt(outfile, data_slice, fmt='%-7.2f')

        # Writing out a break to indicate different slices...
        outfile.write('# New slice\n')
```
