```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

bins = np.linspace(0, 1, 11)
for key in ['AR_R2', 'AR_R3', 'AR_R4', 'AR_R5', 'AR_R6', 'AR_R7', 'AR_R8']:
    for table in ['kmerTableDivide.csv','kmerTableTransit.csv','symmetricKmerTableDivide.csv','symmetricKmerTableTransit.csv']:
        allAffs = []
        firstLine = True
        with open("results/"+key+"/"+table) as inp:
            for line in inp:
                if firstLine:
                    firstLine = False
                    continue
                row = line.strip().split(',')
                if 'symmetric' in table and int(row[2]) == 0: continue
                aff = float(row[-2])
                allAffs.append(aff)
        digitized = np.digitize(allAffs, bins)
        print(key, table)
        for i in range(1, 11): print("bin ", i, " contains ", len(digitized[digitized == i]))
        plt.hist(allAffs, density=False, bins=11)
        plt.ylabel('Counts')
        plt.xlabel('Affinity')
        plt.ylim((0,5.5e7))
        plt.xlim((0,1.01))
        plt.show()
```
