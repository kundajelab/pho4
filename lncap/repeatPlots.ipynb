```python
import numpy as np

ar_a = np.load('/users/amr1/pho4/data/insilico_exp/AR_A/both_mirror_counts.npy')
ar_g = np.load('/users/amr1/pho4/data/insilico_exp/AR_G/both_mirror_counts.npy')
gr_a = np.load('/users/amr1/pho4/data/insilico_exp/GR_A/both_mirror_counts.npy')
gr_g = np.load('/users/amr1/pho4/data/insilico_exp/GR_G/both_mirror_counts.npy')
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 15}

matplotlib.rc('font', **font)
```

```python
plt.plot(ar_a[:10], label = "AR")
frame1 = plt.gca()
frame1.axes.get_yaxis().set_visible(False)
plt.legend()
plt.title("A/T mirrored repeat preferences")
```

```python
plt.plot(ar_a[:10]/np.linalg.norm(ar_a[:10]), label = "AR")
plt.plot(gr_a[:10]/np.linalg.norm(gr_a[:10]), label = "GR")
frame1 = plt.gca()
frame1.axes.get_yaxis().set_visible(False)
plt.legend()
plt.title("A/T mirrored repeat preferences")
```
